```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

stages = pd.Series([1,2,3,4,5])
scenario1 = pd.Series([5,7,6.5,8,10])
scenario2 = pd.Series([4,4.5,5,5.5,6.5])
scenario3 = pd.Series([2,4,6,3.5,5])
```

```python
plt.plot(stages,scenario1,label='Scenario 1',color ='green', linestyle='solid',marker='D')
plt.plot(stages,scenario2,label='Scenario 2',color = 'orange', linestyle = 'dotted', marker = 'o')
plt.plot(stages,scenario3,label='Scenario 3',marker='^',color='purple',linestyle = '-.')
plt.xlabel("Stages")
plt.ylabel("Performance")
plt.legend()
plt.title("Performance Comparison Across Stages")
plt.show()
```

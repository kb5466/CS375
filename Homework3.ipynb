```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
data_path = 'student_grades.csv'
data = pd.read_csv(data_path)

# Split the dataset into features and target variable
X = data.iloc[:, :-1].values  # All rows, all columns except the last one
y = data.iloc[:, -1].values   # All rows, just the last column

# Splitting dataset into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Creating a simple linear regression model
regressor = LinearRegression()

# Training the model
regressor.fit(X_train, y_train)

# Making predictions on the test set
y_pred = regressor.predict(X_test)

# Evaluating the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared score: {r2}")

# Visualizing the training set results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='red', label='Actual')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Predicted')
plt.title('Quiz Grade vs Exam Grade(Training set)')
plt.xlabel('Quiz')
plt.ylabel('Test')
plt.legend()

# Visualizing the test set results
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, color='red', label='Actual')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Prediction Line')
plt.title('Quiz Grade vs Exam Grade(Training set)')
plt.xlabel('Quiz')
plt.ylabel('Test')
plt.legend()
plt.tight_layout()
plt.show()

# Print the model equation
intercept = regressor.intercept_
slope = regressor.coef_[0]
print(f"The linear model is: TestGrade = {slope:.2f} * QuizGrade + {intercept:.2f}")

# Make a prediction for 2.5 years of experience
years_experience = np.array([[80]])
predicted_salary = regressor.predict(years_experience)
print(f"Predicted grade for someone with 80 grade on quiz is: {predicted_salary[0]:.2f}%")
```
